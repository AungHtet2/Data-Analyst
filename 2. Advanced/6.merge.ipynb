```python
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset

# Load Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Objects to Datetime
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_posted_month'] = df['job_posted_date'].dt.month
```

```python
df_usa = df[df['job_country'] == 'United States'].copy()
df_usa['job_posted_month'] = df_usa['job_posted_date'].dt.strftime('%B')
df_usa_pivot_table = df_usa.pivot_table(index='job_posted_month', columns='job_title_short', aggfunc='size')

df_usa_pivot_table = df_usa_pivot_table.reset_index()
df_usa_pivot_table['job_posted_month_no'] = pd.to_datetime(df_usa_pivot_table['job_posted_month'], format= '%B').dt.month 
df_usa_pivot_table = df_usa_pivot_table.sort_values('job_posted_month_no')
df_usa_pivot_table = df_usa_pivot_table.set_index('job_posted_month')
df_usa_pivot_table = df_usa_pivot_table.drop(columns='job_posted_month_no')

df_usa_pivot_table
```

```python
df_software_pivot = pd.read_csv('http://lukeb.co/software_csv',index_col='job_posted_month')
```

```python
df_us_merge = df_usa_pivot_table.merge(df_software_pivot, on='job_posted_month')
```

```python
top_5 = (df_us_merge
         .sum()
         .sort_values(ascending=False)
         .head()
         .index
         .tolist())

top_5
```

```python
df_us_merge[top_5].plot(kind='line')
plt.title('Monthly Job Postings For Top Tech Job in the US')
plt.xlabel('2023')
plt.ylabel('Job Count')
plt.ylim(0,20000)
plt.legend()
plt.show()
```
