```python
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset

# Load Data
dataset = load_dataset('lukebarousse/data_jobs')
df_original = dataset['train'].to_pandas()

# Objects to Datetime
df_original['job_posted_date'] = pd.to_datetime(df_original['job_posted_date'])
df_original['job_posted_month'] = df_original['job_posted_date'].dt.month
```

Copy Method. Didnt alter the orignal file

```python
df_altered = df_original.copy()
```

```python
print('Are the two dataframe same?', id(df_altered) == id(df_original))
```

```python
df_altered.loc[:, 'salary_year_avg' : 'salary_hour_avg']
```

```python
median_year_salary = df_altered['salary_year_avg'].median()
median_hour_salary = df_altered['salary_hour_avg'].median()
```

```python
df_altered['salary_year_avg'] = df_altered['salary_year_avg'].fillna(median_year_salary)
df_altered['salary_hour_avg'] = df_altered['salary_hour_avg'].fillna(median_hour_salary)
```
