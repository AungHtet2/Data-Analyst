```python
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset

# Load Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Objects to Datetime
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

Filter The NaN values in Salary Year Avg

Projected salary for next year

```python
median_year_salary = df['salary_year_avg'].median()
```

```python
df['salary_year_avg'] = df['salary_year_avg'].fillna(median_year_salary)
df.sample(2)
```

```python
def projected_salary(salary):
    return salary * 1.03

df['salary_year_inflation'] = df['salary_year_avg'].apply(projected_salary)

df[['salary_year_avg', 'salary_year_inflation']]
```

Another Method With Lambda

```python
df['salary_year_inflation'] = df['salary_year_avg'].apply(lambda salary: salary * 1.03)

df[['salary_year_avg', 'salary_year_inflation']]
```

```python
def projected_salary(row):
    if 'Senior' in row['job_title_short']:
        return row['salary_year_avg'] * 1.05  # Adjust salary for Senior titles
    else:
        return row['salary_year_avg'] * 1.03  # Adjust salary for others

# Apply the function row-wise
df['salary_year_inflation'] = df.apply(projected_salary, axis=1)
```

```python
df[['job_title_short','salary_year_avg','salary_year_inflation']]
```

USING Lambda

```python
df['salary_year_inflation'] = (df.apply(lambda row: row['salary_year_avg'] * 1.05 if 'Senior' in row['job_title_short'] 
                                        else row['salary_year_avg'] * 1.03, axis=1)
                                        )
```
