```python
# Importing Libraries
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df.head()
```

```python
wfh = df['job_work_from_home'].value_counts()
wfh.plot(kind='pie',startangle=90, autopct="%1.1f%%")
plt.title("Work From Home Status")
plt.ylabel('')
```

```python
job_titles_chart = df['job_title_short'].value_counts()
job_titles_chart.plot(kind='pie', autopct='%1.1f%%')
```

```python
fig, ax = plt.subplots(1,3)
fig.set_size_inches((12, 5))

dict = {
    'job_work_from_home':'Work from Home Offered',
    'job_no_degree_mention' : 'Job Degree Required',
    'job_health_insurance' : 'Health Insurance Offered'
}

for i, (column,title) in enumerate(dict.items()):
    df[column].value_counts().plot(kind='pie', ax=ax[i],startangle=90, autopct="%1.1f%%", labels=['False', 'True'] )
    
    ax[i].set_title(title)
    
```
