```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
csv_path = "D:/kaggle/train.csv"
df = pd.read_csv(csv_path)


```

```python
df.info()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], format='%d/%m/%Y')
df['Ship Date'] = pd.to_datetime(df['Ship Date'], format='%d/%m/%Y')

df['years'] = df['Order Date'].dt.year
```

```python
df.loc[:, 'Product Name' : 'Sales']
```

2018 - Sales

```python
sales_2018 = df[df['Order Date'] == "2018"]
```

```python
sales_2018
```

```python
category_sales_2018 = sales_2018.groupby(['Category'])['Sales'].sum().reset_index()

category_sales_2018.plot(kind='bar', x='Category', y='Sales')
plt.title('Sales by Category')
plt.xlabel('Category')
plt.ylabel('Total sales')
plt.xticks(rotation = 0)
plt.show()
```

Highest sales of 2018

```python
highest_sales_2018 = sales_2018['Sales'].idxmax()
df.iloc[highest_sales_2018]
```

Most buy Products of 2018

```python
most_buy_products_2018 = sales_2018['Product Name'].idxmax()
most_buy_products_2018 = df.iloc[most_buy_products_2018]
print(f'{sales_2018['Product Name'].max()} is a most buy products of 2018.')
most_buy_products_2018
```

"States' most purchased products of 2018."

```python
state_most_purchased = sales_2018.groupby(['State'])['Sales'].sum().reset_index()


state_most_purchased.plot(kind='bar', x='State', y='Sales')
plt.title('"States" most purchased products of 2018.')
plt.xlabel('States')
plt.ylabel('Total Sales')
plt.xticks(rotation=0)
plt.show()
```

"City" most purchased products from Texas 2018

```python
city_most_purchased = sales_2018.groupby(['City'])['Sales'].sum().reset_index()
city_most_purchased
```

```python
city_most_purchased.plot(kind='barh', x='City', y='Sales')
plt.title("'City' most purchased products from Texas 2018")
plt.ylabel('Cities')
plt.xlabel('Total Sales')
plt.show()
```

```python
top_states = df['State'].value_counts().head(6).index
```

```python
df_top_states_purchased = df.pivot_table(values='Sales',
                                          index='State',
                                            columns='Category',
                                              aggfunc='median')
df_top_states_purchased = df_top_states_purchased.loc[top_states]
df_top_states_purchased = df_top_states_purchased[['Furniture', 'Technology', 'Office Supplies']]


df_top_states_purchased.plot(kind='bar')
plt.title('Median purchased by States and Category')
plt.xlabel('Top States')
plt.ylabel('Median purchased(USD$)')
plt.xticks(rotation=45)

# Move legend outside
plt.legend(title="Category", bbox_to_anchor=(1.05, 1), loc='upper left')

plt.show()
```
