```python
top3_populations = df.sort_values(by='2022 Population',ascending=False).head(5)
top3_populations
```

```python
sns.set_theme(style='dark')
ax = sns.barplot(data=top3_populations, x='Country/Territory', y='2022 Population',hue='Country/Territory', palette='viridis')
plt.title('Top 5 Most Populated Country in 2022')
plt.xlabel('Country')


# Format y-axis labels to show billions (B) and millions (M)
def format_population(x, pos):
    if x >= 1_000_000_000:  # Convert to billions
        return f'{x / 1_000_000_000:.1f}B'
    elif x >= 1_000_000:  # Convert to millions
        return f'{x / 1_000_000:.1f}M'
    return f'{x:.0f}'

ax.yaxis.set_major_formatter(ticker.FuncFormatter(format_population))


plt.tight_layout()
plt.show()
```

```python
df_area_and_density = (df.groupby('Country/Territory')[['Area (km²)','Density (per km²)']]
                       .max()
                       .sort_values(by='Area (km²)',ascending=False)
                       .head(5))
df_area_and_density
```

```python
# Set the Seaborn theme for better visualization
sns.set_theme(style="whitegrid")

# Create a figure with one axis
fig, ax1 = plt.subplots(figsize=(12, 8))

# Plot the bar chart for Area
sns.barplot(x=df_area_and_density.index, y=df_area_and_density['Area (km²)'], ax=ax1, color='lightblue')

# Create a second axis to plot the line chart for Density
ax2 = ax1.twinx()  # Share the same x-axis

# Plot the smooth line for Density (optional: you can use smoothing if needed)
sns.lineplot(x=df_area_and_density.index, y=df_area_and_density['Density (per km²)'], ax=ax2, color='red', linewidth=2)

# Set labels for the axes
ax1.set_xlabel('Country/Territory')
ax1.set_ylabel('Area (km²)', color='blue')
ax2.set_ylabel('Density (per km²)', color='red')

# Rotate x-axis labels for readability
plt.xticks(rotation=90)

# Set title for the plot
plt.title('Area and Density of Countries/Territories')

# Display the plot
plt.tight_layout()
plt.show()
```
