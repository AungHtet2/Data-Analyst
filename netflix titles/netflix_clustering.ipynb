```python
import pandas as pd
import matplotlib.pyplot as plt
import ast
```

```python
df = pd.read_csv(r'D:\data analyst\netflix titles\netflix_titles.csv')


df['date_added'] = pd.to_datetime(df['date_added'], format='mixed')
df['release_year'] = df['release_year'].astype(int)
```

```python
df.head()
```

```python
df.tail()
```

```python
df.isna().sum()
```

```python
df = df.dropna(subset=['director','cast','country'])
```

```python
df
```

```python
df_movie = df[df['type'] == 'Movie'].copy()
df_show = df[df['type'] == 'TV Show'].copy()
```

```python
genres_list = df['listed_in'].to_list()
df['listed_in'] = genres_list
```

```python
type(df['listed_in'][7])
```

```python
df['listed_in'] = df['listed_in'].apply(lambda genre: genre.split(', '))
type(df['listed_in'][7])
```

```python

genre_explode = df.explode('listed_in')
genre_explode_movie = genre_explode[genre_explode['type'] == 'Movie']
df_genre_movie_counts = (genre_explode_movie['listed_in'].value_counts()
                         .reset_index(name='Genre Counts')
                         .sort_values(by='Genre Counts',ascending=True)
                         )

plt.figure(figsize=(10, 6))
df_genre_movie_counts.head(10).plot(kind='barh',x='listed_in',y='Genre Counts')
plt.xlabel('Listed Genre')
plt.ylabel('Genre Counts')
plt.title('Top 10 Movie Genres on Netflix')
#plt.xticks(rotation=45)
plt.legend().set_visible(False)
plt.show()
```

```python
genre_explode = df.explode('listed_in')
genre_explode_shows = genre_explode[genre_explode['type'] == 'TV Show']
df_genre_shows_counts = (genre_explode_shows['listed_in'].value_counts()
                         .reset_index(name='Genre Counts')
                         .sort_values(by='Genre Counts',ascending=True)
                         )

plt.figure(figsize=(10, 6))
df_genre_shows_counts.head(10).plot(kind='barh',x='listed_in',y='Genre Counts')
plt.xlabel('Listed Genre')
plt.ylabel('Genre Counts')
plt.title('Top 10 TV Show Genres on Netflix')
#plt.xticks(rotation=45)
plt.legend().set_visible(False)
plt.show()
```

```python
df_movie_country_counts = df_movie['country'].value_counts().head(5).sort_values(ascending=True)

plt.figure(figsize=(10, 6))
df_movie_country_counts.plot(kind='barh')
plt.xlabel('Country Count')
plt.ylabel('Country')
plt.title('Top 5 Movie Countries on Netflix')
plt.show()
```

```python
df_show_country_counts = df_show['country'].value_counts().head(5).sort_values(ascending=True)

plt.figure(figsize=(10, 6))
df_show_country_counts.plot(kind='barh')
plt.xlabel('Country Count')
plt.ylabel('Country')
plt.title('Top 5 TV Shows Countries on Netflix')
plt.show()
```
